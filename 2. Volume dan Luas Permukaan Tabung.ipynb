### Problem 2 (Cylinder)
Coba buatlah method/fungsi yang berisi formulasi dari volume dan luas tabung yang mana keduanya di simpan di dalam suatu class bernama 'Cylinder'

`HINT`

**VOLUME OF CYLINDER**


**SURFACE AREA OF CYLINDER**


Source:
- https://www.omnicalculator.com/math/cylinder-volume
- https://study.com/academy/lesson/finding-the-area-of-a-cylinder-formula-example.html

````python
class Cylinder:
    pi = 3.14
    
    def __init__ (self, height = 1, radius = 1): #by default kalo lo gak memberikan nilai, dia akan menggunakan nilai 'height=1' dan 'radius = 1'
        self.height = height
        self.radius = radius
    
    #Equation method of cylinder volume
    def volume (self):
        return self.pi * (self.radius**2) * self.height 
    
    #Equation method of the surface area of the cylinder
    def surface (self):
        return (2*self.pi * self.radius) * (self.radius+self.height)
````

````python
#EXAMPLE OUTPUT
c = Cylinder(2,3)
````

````python
c.volume()
````

````python
c.surface()
````

```
Jadi, pada 'Problem 2' kali ini tuh lo di suruh untuk membuat class 'Cylinder' yang di dalamnya terdapat formulasi perhitungan 'volume' dan 'surface'

class Cylinder:
    pi = 3.14
    
    def __init__ (self, height = 1, radius = 1): 
        self.height = height
        self.radius = radius

by default, kalo lo gak memberikan nilai pada argument di dalam 'Magic/Dunder' method '__init__', dia akan menggunakan nilai 'height=1' dan 'radius = 1' saat dieksekusi.

Untuk membuat formulasi perhitungan, lo bisa menggunakan teknik yang ada di dalam class, yakni 'method' di dalam OOP atau gampangnya mah 'function'

'method = function'

`ALUR LOGIKA DALAM MEMBUAT OBJECT PADA PROBLEM 2`
1. Problemnya tuh lo disuruh membuat formulasi untuk menghitung volume dan luas permukaan tabung (Cylinder).

2. Kalo dilihat rumusnya, maka lo membutuhkan nilai 'pi, tinggi (height), dan jari-jari (radius)'

3. Karena membutuhkan ketiga nilai itu, maka argument di dalam 'Magic/Dunder' methodnya itu merupakan object untuk nilai dari masing-masing rumusnya, yakni 'height' dan 'radius'

`Pertanyaannya:` 
Kenapa nilai 'pi' berada di luar method/function? dan kenapa menggunakan 'self' pada 'pi'-nya?

'Jawabannya:'
Nilai 'pi' pada kasus ini diletakkan sebagai 'Global variable' yang bernilai 3.14. 'Global variable' merupakan suatu variable yang diletakkan diluar method/function sehingga untuk mengakses nilai 'pi'-nya di dalam method, LO HARUS MENGGUNAKAN 'self' UNTUK MENDEFINISIKAN NILAI 'pi' NYA ITU SENDIRI. 

Hal ini karena argument di dalam 'method'-nya itu adalah 'self' sehingga semua perhitungan yang ada di dalam methodnya itu HARUS MENGGUNAKAN 'self'
```
