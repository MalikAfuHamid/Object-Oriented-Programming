````python
class Line:
    
    def __init__ (self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2
    
    #DISTANCE FORMULA
    def distance(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        return ((x2-x1)**2 + (y2-y1)**2)**0.5
    
    #SLOPE FORMULA
    def slope(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        return (y2-y1) / (x2-x1)
````

````python
#EXAMPLE OUTPUT

coordinate1 = (3,2) #x1, y1
coordinate2 = (8,10) #x2, y2

li = Line(coordinate1, coordinate2)
````

````python
import numpy as np
````

````python
class Line:
    
    def __init__ (self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def distance(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        
        return np.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    def slope (self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        
        return (y2-y1) / (x2 - x1)
````

````python
c1 = (3,2)
c2 = (8,10)

li = Line(c1, c2)
````

````python
li.distance()
````

````python
li.slope()
````

```
Jadi gua baru nemuin nih, ternyata, kalo nilai arugment 'self' di dalam method/fungsi itu diganti dengan kata lain, LO TETAP BISA MENGEKSEKUSI HASILNYA.

The question is 'WHY DID IT HAPPEN?' ALWAYS TRY TO ASK WHY!!!!!

The answer is:
Jadi gini, 'self' di dalam '__init__' itu merupakan metode OOP yang digunakan untuk lo bisa membuat object tertentu yang nantinya 'object' tersebut akan dijadikan suatu 'attribute' yang bisa diakses. Jadi dengan menggunakan argument 'self' di dalam '__init__', lo bisa mengakses kembali object yang lo mau kapanpun itu.

'FYI' --> Sebenarnya, namanya itu gak harus 'self' karena 'self' sendiri hanyalah argument yang bisa dibuat sesuka lo mau apa namanya. Akan tetapi, karena ini sudah kesepakatan bersama antar programmer di seluruh dunia, maka biar gampang pahamnya pake 'self' aja, ya!

Argument di dalam method/fungsi itu sebenarnya gak harus 'self' — lo bisa menuliskan nama argumentnya apa saja. Hal ini terjadi karena sebenarnya antara '__init__' dan 'method' itu 2 hal yang terpisah sehingga walaupun nama yang lo tuliskan itu bukan 'self' di dalam 'method'-nya, maka lo tetap bisa mengeksekusi hasilnya sebagai suatu hasil yang berasal dari fungsi (def) biasa.

Akan tetapi, kalo lo tidak menuliskan 'self' seperti yang ada di dalam 'Magic/Dunder' methodnya, maka object yang lo buat itu tidak saling terkoneksi. Hal ini menyebabkan lo tidak bisa mengakses sembarangan lagi terhadap object yang udah lo buat.
```
